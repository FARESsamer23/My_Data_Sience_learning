### Fonction

Arguments *ARGS ET **KWARGS DANS LES FONCTIONS

 - **kwargs dans les définitions de fonctions en python permet de transmettre une liste d'arguments de longueur variable avec mots-clés.
-  Lorsque vous passez un argument portant un nom à la fonction, nous l'appelons un mot clé.
- L’ordre dans lequel les noms d’arguments a été ajouté n’est pas important.


```python
def Info(**kwargs):
    print('votre nom est', kwargs['nom'])
    print('votre prénom est', kwargs['prenom'])
    print('ville',kwargs['ville'])
    #print('age',kwargs['age'])

Info(nom='Nedjla',prenom='Ward', ville='Mila', age=31)

Info(nom='yacine',prenom='hu', ville='Mila')
u=['u',87]
Info(u)
```

```python

def fonction(**kwargs):
    for key, value in kwargs.items():
        print("%s == %s" % (key, value))


fonction(prenom='Nedjla', nom='Ward', ville='Mila')

fonction(prenom='Info', nom='Blida')
```

```python
def fonction(arg1, **kwargs):
    #print(arg1)
    if arg1 ==31:
        for key, value in kwargs.items():
            print("%s == %s" % (key, value))


fonction(31, prenom='Nedjla', nom='Ward', ville='Mila')
```

- L'argument spécial *args dans la définition d'une fonction en python permet de transmettre une liste d'arguments de longueur variable sans mot-clé à une fonction.
- *args vous permet de prendre plus d'arguments que le nombre d'arguments formels prédéfinis.
- En utilisant *, la variable que nous associons à * devient un mot éditable, ce qui signifie que vous pouvez effectuer des opérations comme itérer dessus.



```python
def fonction(*args):
    for elm in args:
        print(elm)

fonction(2,'Nedjla', 31, [1,3])
```

```python
def Info(nom, *args):
    print('votre nom est', nom)
    print('votre prénom est', args[0])
    print('votre ville est', args[1])
    print('votre age est', args[2])
Info('Nejdla', 'Ward', 'Mila', 31)
```

Il est possible d’utiliser les mots clés *args et **kwargs en même temps.


```python
def Info(*args, **kwargs):
    print('votre nom est', args[0])
    print('votre prénom est', kwargs['prenom'])
    print('ville',kwargs['ville'])
    print('age',kwargs['age'])

Info('Nedjla',prenom='Ward', ville='Mila', age=31)
```

Typage de paramètres
En Python, le typage des paramètres d'une fonction est un typage dynamique, ce qui signifie que le type d\'un paramètre devient automatiquement le même que celui de l'argument qui a été transmis ) la fonction

```python
def afficher3fois(param):
    print(param, param, param)

afficher3fois(5) #int
afficher3fois('Hi ') #string
afficher3fois([1,5]) #liste
afficher3fois(6**3) #expression
```

#### Arguments avec étiquettes
Dans la plupart des langages de programmation, les arguments que l’on fournit lors de l’appel d’une
fonction doivent être fournis exactement dans le même ordre que celui des paramètres qui leur correspondent dans la définition de la fonction.
Python autorise cependant une souplesse beaucoup plus grande. Si les paramètres annoncés dans la définition de la fonction ont reçu chacun une valeur par défaut, sous la forme déjà décrite ci-dessus, on peut faire appel à la fonction en fournissant les arguments correspondants dans n’importe quel ordre, à la condition de désigner nommément les paramètres correspondants. Exemple :

```python
def oiseau(voltage=100, etat='fou', action='danser la java'):
    print('Ce perroquet ne pourra pas', action)
    print('si vous le branchez sur', voltage, 'volts !')
    print("L'auteur de ceci est complètement", etat)

oiseau(etat='givré', voltage=250, action='vous approuver')

oiseau()
```

### Fonction de haut niveau
Nous pouvons réaliser des programmes expressifs et succincts en utilisant des fonctions de haut niveau. Ces dernières acceptent une fonction comme argument ou renvoient une fonction comme valeur.
On peut utiliser une fonction de haut niveau comme moyen de créer des fonctions composites à partir de fonctions plus simples. Exemple:


```python
year_cheese = [(2000, 29.87), (2001, 30.12), (2002, 30.6), (2003,
30.66),(2004, 31.33), (2005, 32.62), (2006, 32.73),(2007, 33.5),
(2008, 35), (2009, 40), (2010, 32.92)]

max(year_cheese)
 #Par défaut retourne la plus grande valeur dans la position 0
```

```python
max(year_cheese, key=lambda yc: yc[1])
  #Par défaut retourne la plus grande valeur dans la position 1
```

```python
import numpy as np
import scipy as sp
print(np.sqrt(-1))
print(sp.sqrt(-1))
```
