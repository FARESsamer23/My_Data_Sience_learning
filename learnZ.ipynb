```python
class Employee:
    company = "Google"  # class variable

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

e1 = Employee("Samer", 5000)
e2 = Employee("Sara", 7000)

print(e1.company)  # Google
print(e2.company)  # Google

Employee.company = "OpenAI"
print(e1.company)  # OpenAI
```
