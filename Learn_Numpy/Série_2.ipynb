7. Écrire une fonction compterMots ayant un argument (une chaîne de caractères) qui renvoie un
dictionnaire qui contient la fréquence de tous les mots de la chaîne entrée.

```python
def compterMots(chaine):
    # Supprimer la ponctuation et convertir en minuscules
    import string
    chaine = chaine.translate(str.maketrans('', '', string.punctuation)).lower()
    
    # Diviser la chaîne en mots
    mots = chaine.split()
    
    # Créer un dictionnaire pour compter les fréquences
    frequence = {}
    for mot in mots:
        if mot in frequence:
            frequence[mot] += 1
        else:
            frequence[mot] = 1
    
    return frequence

# Exemple d'utilisation
texte = "Bonjour, bonjour! Comment ça va? Ça va bien, merci."
resultat = compterMots(texte)
print(resultat)
      
```

```python
def trouver_elements_repetes(mon_tuple):
    # Créer un dictionnaire pour compter les occurrences
    compteur = {}
    for element in mon_tuple:
        if element in compteur:
            compteur[element] += 1
        else:
            compteur[element] = 1
    
    # Filtrer les éléments répétés
    elements_repetes = [key for key, value in compteur.items() if value > 1]
    return elements_repetes

# Exemple d'utilisation
mon_tuple = (1, 2, 3, 4, 5, 2, 3, 6, 7, 3, 8, 1)
resultat = trouver_elements_repetes(mon_tuple)
print("Éléments répétés :", resultat)
```

```python
def verifier_element_existe(mon_tuple, element):
    # Vérifier si l'élément est dans le tuple
    return element in mon_tuple

# Exemple d'utilisation
mon_tuple = (10, 20, 30, 40, 50)
element = 30

if verifier_element_existe(mon_tuple, element):
    print(f"L'élément {element} existe dans le tuple.")
else:
    print(f"L'élément {element} n'existe pas dans le tuple.")
```

```python
# Définir les ensembles
X = {'a', 'b', 'c', 'd'}
Y = {'s', 'b', 'd'}

# Afficher les ensembles initiaux
print("Ensemble X :", X)
print("Ensemble Y :", Y)

# Test d'appartenance
print("L'élément 'c' appartient à X :", 'c' in X)
print("L'élément 'a' appartient à Y :", 'a' in Y)

# Calcul des différences
print("X - Y :", X - Y)
print("Y - X :", Y - X)

# Calcul de l'union
print("X union Y :", X | Y)

# Calcul de l'intersection
print("X intersection Y :", X & Y)
```

```python
import math

# Fonction de soustraction
def soustraction(a, b):
    resultat = a - b
    racine = math.sqrt(abs(resultat))  # Racine carrée de la valeur absolue
    return (resultat, racine)

# Fonction d'addition
def addition(a, b):
    resultat = a + b
    racine = math.sqrt(resultat)
    return (resultat, racine)

# Fonction de multiplication (accepte un nombre arbitraire d'arguments)
def multiplication(*args):
    resultat = 1
    for arg in args:
        resultat *= arg
    racine = math.sqrt(resultat)
    return (resultat, racine)

# Tests avec plusieurs jeux de valeurs
print("Test de soustraction :")
print(soustraction(10, 4))  
print(soustraction(4, 10))  

print("\nTest d'addition :")
print(addition(10, 4))  
print(addition(5, 5))   

print("\nTest de multiplication :")
print(multiplication(2, 3))          
print(multiplication(2, 3, 4))       
print(multiplication(1, 2, 3, 4, 5)) 
```
